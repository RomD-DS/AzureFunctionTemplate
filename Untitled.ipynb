```python
import msal
import requests

# Mets ici tes valeurs depuis App Registration
CLIENT_ID = "c8280fc0-7bad-4373-8e7e-da3567255da4"
TENANT_ID = "8b87af7d-8647-4dc7-8df4-5f69a2011bb5"

# Permission minimale
SCOPES = ["Mail.Read"]

# 1) App MSAL
app = msal.PublicClientApplication(
    client_id=CLIENT_ID,
    authority=f"https://login.microsoftonline.com/{TENANT_ID}"
)
```

```python

# 2) Authentification Device Code (super simple)
flow = app.initiate_device_flow(scopes=SCOPES)
print(flow["message"])      # Tu cliques sur le lien et acceptes

result = app.acquire_token_by_device_flow(flow)

if "access_token" not in result:
    raise Exception("Impossible de se connecter :", result)

token = result["access_token"]

# 3) Lire 1 mail
url = "https://graph.microsoft.com/v1.0/me/messages?$top=1"
headers = {"Authorization": f"Bearer {token}"}

res = requests.get(url, headers=headers).json()

mail = res["value"][0]
print("\n===== MAIL =====")
print("From :", mail["from"]["emailAddress"]["address"])
print("Subject :", mail["subject"])
print("Preview :", mail["bodyPreview"])
```
